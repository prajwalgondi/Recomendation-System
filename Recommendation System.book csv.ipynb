```python
import pandas as pd
import numpy as np
from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine,correlation
```

```python
# Import Dataset
df=pd.read_csv('book (1).csv',encoding=('ISO-8859-1'),low_memory=False)
df
```

```python
df1=df.iloc[:,1:]
df1
```

```python
###sort by user id
```

```python
df1.sort_values('User.ID')
```

```python
###unique users in dataset
```

```python
len(df1['User.ID'].unique())
```

```python
len(df1['Book.Title'].unique())
```

```python
###converting data into pivot table
```

```python
df2=df1.pivot_table(index='User.ID',columns='Book.Title',values='Book.Rating').reset_index(drop=True)
df2
```

```python
####replacing index values to unique values
```

```python
df2.index=df1['User.ID'].unique()
df2
```

```python
df2.fillna(0,inplace=True)
df2
```

```python
###cosine similarity
```

```python
user_sim=1-pairwise_distances(df2.values,metric='cosine')
user_sim
```

```python
user_sim1=pd.DataFrame(user_sim)
user_sim1
```

```python
####index and columns names to user id
```

```python
user_sim1.index=df1['User.ID'].unique()
user_sim1.columns=df1['User.ID'].unique()
user_sim1
```

```python
###null the diagonal vlaues
```

```python
np.fill_diagonal(user_sim,0)
user_sim1
```

```python
###similar values
```

```python
user_sim1.idxmax(axis=1)
```
